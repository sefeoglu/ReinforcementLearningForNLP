```python
import pandas as pd
import json
import os
import numpy
```

```python
path = os.path.join("../data/pure/", "dataset_chitchat.py")
output = pd.DataFrame()
dataset = json.load(open(path))
```

```python
id_list = []
for convo_id, convo in dataset.items():
    id_list.append(convo_id)
len(id_list)
```

```python
output = pd.DataFrame()
dataset = json.load(open(path))
first = id_list[5000]

start_flag = False
for convo_id, convo in dataset.items():
    
    if first == convo_id:
        start_flag = True
        
    if start_flag:
        if 'prompt' in convo:
            prompt = convo['prompt']
        else:
            prompt = ""
        if 'start' in convo:
            start = convo['start']
        else:
            start = ""
        if 'ratings' in convo:
            ratings = convo['ratings']
        else:
            ratings = ""
        for a in convo['messages']:
            l = list(a)
            text = ''
            for m in l:
                text += m['text']
                text = text +" / "
            message_data = dict({'conv_id':convo_id,'prompt':prompt,'start':start,'ratings':ratings,'text': text, 'sender': m['sender'], 'timestamp':  m['timestamp']})
            output = output.append(message_data, ignore_index=True)   
output
```
